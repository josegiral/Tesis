## Modulos de Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definición de Funciones

```python
def OrbitasSistema(f,x0, n, l, u, ax=None): 
    """
    Función que grafica las orbitas del sistema dinámico discreto
    dado por x+1 = f(x) utilizando la semilla x0. Los parámetros 
    que resive esta función son:

        f  : una función que define el sistema dinámico.
    
        x0 : un flotante que sirve como semilla para la orbita.
    
        n  : un entero que determina el número de iteraciones.
    
        l  : un flotante que sirve como límite inferior de los ejes.
    
        u  : un flotante que sirve como límite superior de los ejes.
    
        ax : un sistema de ejes para graficar.
    
    La función regresa las gráficas de las órbitas en en el sistema de ejes dado.
    """
    
    t = np.linspace(l, u)
    ax.plot(t, f(t), 'b', lw=2, label = r'$\phi(t)= e^{2(t-1)}$')
    ax.plot([l, u], [l, u], 'k', lw=2)
    
    # Recursively apply y=f(x) and plot two lines:
    # (x, x) -> (x, y)
    # (x, y) -> (y, y)
    x = x0
    for i in range(n):
        
        y = f(x)
        
        # Plot the two lines.
        ax.plot([x, x], [x, y], 'r', lw=1)
        ax.plot([x, y], [y, y], 'r', lw=1)
        
        # Plot the positions with increasing
        # opacity.
        ax.plot([x], [y], 'or', ms=5, alpha=(i + 1) / n)
        x = y
        
    ax.plot([x], [y], 'or', ms=5, alpha=(i + 1) / n, label='x = '+str(x))

    ax.set_xlim(l, u)
    ax.set_ylim(l, u)
    ax.legend()
```

## Cobweb

```python
fig,(ax1) = plt.subplots(1, 1, figsize=(7,7))

Q = lambda x : np.exp(2*(x-1))

x1 = 0

OrbitasSistema(Q, x1, 100, 0, 1, ax =ax1)
```
